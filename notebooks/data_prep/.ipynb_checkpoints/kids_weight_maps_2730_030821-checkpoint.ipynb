```python
hdul = fits.open('/home/shawn/Downloads/Sci-ADVORNIK-OMEGACAM-------OCAM_r_SDSS---Coadd-GAstrom-Sci-58202.4677137-c8b1bef74c0042bf292d17df5e303af50ba7020a.fits')
coadd_header = hdul[0].header
coadd_image = hdul[0].data
hdul.close()

print(coadd_image)
print('median pixel: ', np.median(coadd_image))
print(coadd_header)
```
