# 09/12/20 - Changing size of image and noise_map.

```python
# libraries

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from astropy.io import fits
import autolens as al
import autolens.plot as aplt
```

```python
# set datapaths
links_path = '/home/shawn/Desktop/links_autoz/'
fits_path = f'{links_path}files/fits/'
fits_image_path = f'{fits_path}images/'
csv_path = f'{links_path}files/csv/'
png_path = f'{links_path}visuals/png/'
pdf_path = f'{links_path}visuals/pdf/'
```

```python
# load in r-band image as an array
image = al.Array.from_fits(
    file_path=f'{fits_image_path}links/G3629152_1933/1933_r.fits',
    pixel_scales=0.198
)

aplt.Array(image)
```

```python
# resize to smaller shape. (51x51 gives about 10 x 10 arcsec)
zoom_image = al.preprocess.array_with_new_shape(
    array=image, new_shape=(51, 51)
)

aplt.Array(array=zoom_image)
```

```python
noise_map = al.Array.from_fits(
    file_path=f'{fits_image_path}links/G3629152_1933/1933_r_noisemap.fits',
    pixel_scales=0.198
)

aplt.Array(noise_map)

# resize to smaller shape. (51x51 gives about 10 x 10 arcsec)
zoom_noise_map = al.preprocess.array_with_new_shape(
    array=noise_map, new_shape=(51, 51)
)

aplt.Array(array=zoom_noise_map)
```
