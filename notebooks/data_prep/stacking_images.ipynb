# 09/12/20 - Stack g, r, and i images.

```python
# libraries

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from astropy.io import fits
import autolens as al
import autolens.plot as aplt

# set datapaths
links_path = '/home/shawn/Desktop/links_autoz/'
fits_path = f'{links_path}files/fits/'
fits_image_path = f'{fits_path}images/'
csv_path = f'{links_path}files/csv/'
png_path = f'{links_path}visuals/png/'
pdf_path = f'{links_path}visuals/pdf/'
```

```python
# import g, r, and i images
# load in g-band image as an array
g_image = al.Array.from_fits(
    file_path=f'{fits_image_path}links/G3629152_1933/1933_g.fits',
    pixel_scales=0.198
)
g_image_trimmed = al.preprocess.array_with_new_shape(
    array=g_image, new_shape=(51, 51)
)

aplt.Array(g_image_trimmed)

# load in r-band image as an array
r_image = al.Array.from_fits(
    file_path=f'{fits_image_path}links/G3629152_1933/1933_r.fits',
    pixel_scales=0.198
)
r_image_trimmed = al.preprocess.array_with_new_shape(
    array=r_image, new_shape=(51, 51)
)

aplt.Array(r_image_trimmed)

# load in i-band image as an array
i_image = al.Array.from_fits(
    file_path=f'{fits_image_path}links/G3629152_1933/1933_i.fits',
    pixel_scales=0.198
)
i_image_trimmed = al.preprocess.array_with_new_shape(
    array=i_image, new_shape=(51, 51)
)

aplt.Array(i_image_trimmed)
```

```python
# sum arrays element-wise
stacked_image = g_image_trimmed + r_image_trimmed + i_image_trimmed

aplt.Array(stacked_image)
```
