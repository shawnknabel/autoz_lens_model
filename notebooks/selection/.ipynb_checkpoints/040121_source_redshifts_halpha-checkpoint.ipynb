# 04/01/21 - This notebook shows h-alpha line (6562 nm) redshifted from source. ie. 
# $\lambda=\lambda_0\times(1+z)$ where $\lambda_0 = 6562$ nm

```python
%matplotlib inline
import numpy as np
import pandas as pd
from astropy.io import fits
import matplotlib.pyplot as plt

from pyprojroot import here

autoz_path = str(here())
file_path = f'{autoz_path}/files/'
csv_path = f'{file_path}/csv/'
fits_path = f'{file_path}/fits/'
visuals_path = f'{autoz_path}/visuals/'
png_path = f'{visuals_path}/png/'
pdf_path = f'{visuals_path}/pdf/'
```

```python
# load sample csv
samples = pd.read_csv(f'{csv_path}latest/all_samples_latest.csv')
print(samples.columns)
```

```python
plt.hist(samples.zsource, bins=12)
plt.title('Source Redshifts')
plt.xlabel('z')
plt.xlim(0.3, 0.9)
plt.ylabel('Count')
```

```python
lambda_halpha = 6562 # nm
samples['halpha_source'] = lambda_halpha * (1 + samples['zsource'])
```

```python
# plot figure
halpha = samples['halpha_source'].sort_values(ascending=True)
gama_ids = samples['GAMA_ID'].astype('int').astype('str')

fig, axs = plt.subplots(figsize=(16,6))
axs.set_xlim(8500, 12500)
axs.set_ylim(0, 0.3)
axs.set_xlabel('Wavelenght (Angstroms)')
axs.set_ylabel('Transmission')

lambdas = []

for i in range(len(halpha)):
    x = halpha[i]
    ymax = 0.65
    lambdas_diff = lambdas - x
    lambdas_close = lambdas_diff[abs(lambdas_diff) < 100]
    ymax = ymax - 0.1*len(lambdas_close)
    axs.axvline(x=halpha[i],
                ymin=0,
                ymax=ymax,
               color='k',
               linewidth='0.75')
        # the x coords of this transformation are data, and the
        # y coord are axes
    lambdas.append(x)
    trans = axs.get_xaxis_transform()
    plt.text(x=halpha[i]-10, y=ymax+0.02, s=f'G{gama_ids[i]}', rotation='vertical', transform=trans, fontsize=7)
    
print(lambdas)
```

```python
# load in filters
hst_814w = np.genfromtxt(f'{file_path}other/hst_filters/HST_WFC3_UVIS1_F814W_filter.txt', delimiter='')
#print(hst_814w)
hst_105w = np.genfromtxt(f'{file_path}other/hst_filters/HST_WFC3_IR_F105W_filter.txt', delimiter='')
hst_125w = np.genfromtxt(f'{file_path}other/hst_filters/HST_WFC3_IR_F125W_filter.txt', delimiter='')
hst_110w = np.genfromtxt(f'{file_path}other/hst_filters/HST_WFC3_IR_F110W_filter.txt', delimiter='')
hst_160w = np.genfromtxt(f'{file_path}other/hst_filters/HST_WFC3_IR_F160W_filter.txt', delimiter='')
```

```python
# plot figure
halpha = samples['halpha_source'].sort_values(ascending=True)
gama_ids = samples['GAMA_ID'].astype('int').astype('str')

fig, axs = plt.subplots(figsize=(16,6))
axs.set_xlim(8500, 12500)
axs.set_ylim(0, 0.6)
axs.set_xlabel('Wavelength (Angstroms)')
axs.set_ylabel('Transmission')
axs.set_title('Background Sources H-$\\alpha$ Line Redshifted to HST IR Filters')

axs.plot(hst_814w[:,0], hst_814w[:,1],
        color='b',
        label='WFC3 UVIS 814W')
axs.plot(hst_105w[:,0], hst_105w[:,1],
        color='r',
        label='WFC3 IR 105W')
axs.plot(hst_125w[:,0], hst_125w[:,1],
        color='g',
        label='WFC3 IR 125W')
axs.plot(hst_110w[:,0], hst_110w[:,1],
        color='purple',
        label='WFC3 IR 110W')
axs.plot(hst_160w[:,0], hst_160w[:,1],
        color='gold',
        label='WFC3 IR 160W')
axs.legend()

lambdas = []

for i in range(len(halpha)):
    x = halpha[i]
    ymax = 0.65
    lambdas_diff = lambdas - x
    lambdas_close = lambdas_diff[abs(lambdas_diff) < 150]
    ymax = ymax - 0.1*len(lambdas_close)
    axs.axvline(x=halpha[i],
                ymin=0,
                ymax=ymax,
               color='k',
               linewidth='0.75')
        # the x coords of this transformation are data, and the
        # y coord are axes
    lambdas.append(x)
    trans = axs.get_xaxis_transform()
    plt.text(x=halpha[i]-5, y=ymax+0.02, s=f'G{gama_ids[i]}', rotation='vertical', transform=trans, fontsize=7)

plt.savefig(f'{png_path}source_halpha_hst_filters_narrow.png')
#print(lambdas)
```
