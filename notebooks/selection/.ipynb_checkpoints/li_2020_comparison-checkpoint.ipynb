```python
# match li to gama via RA/DEC to get GAMA_IDs


li_in_gama['GAMA_ID'] = 0

for i in range(len(li_in_gama)):
    print('Li candidate', i, 'processing...')
    r_close = 1
    rlim = 2./3600
    gama_id_match = 0
    for j in range(len(gama)):
        x = li_in_gama.RA[i]
        X = gama.RA[j]
        y = li_in_gama.DEC[i]
        Y = gama.DEC[j]
        r = np.sqrt((x - X)**2 + (y - Y)**2)
        if r_close < rlim:
            rlim = rclose
        if r < rlim:
            rclose = r
            gama_id_match = gama.GAMA_ID[j]
            print('Closest match:', gama_id_match, rclose)
    li_in_gama['GAMA_ID'][i] = gama_id_match
    if li_in_gama['GAMA_ID'][i] == 0:
        print('No match.')
    else:
        print('Li candidate', i, 'matches to', gama_id_match)
        
print('Finished.')
        
            
```
