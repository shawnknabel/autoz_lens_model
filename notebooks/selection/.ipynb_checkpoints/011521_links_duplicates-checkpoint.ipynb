# 1/15/21 - This notebook examines the final duplicates of the LinKS catalog to pare it down to the final choices.

```python
# libraries
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
%matplotlib inline

# paths
csv_path = '/data/sknabel/autoz_lens_model/files/csv/'
latest_path= f'{csv_path}latest/'
png_path = '/data/sknabel/autoz_lens_model/visuals/png/'
```

```python
# load in links and links-knabel
links = pd.read_csv(f'{latest_path}links_sample_latest_len42.csv')
links_knabel = pd.read_csv(f'{latest_path}links_knabel_sample_latest_len7.csv')

print(len(links), len(links_knabel))
```

```python
# examine duplicates in links
links_duplicates = links[links.duplicated(subset=['GAMA_ID'], keep=False)]
links_duplicates
```

```python
links_duplicates[['GAMA_ID', 'type', 'zlens', 'sigma_lens', 'zsource', 'sigma_source', 'Z', 'template1', 'template2']]
```

### G544226 has two entries.
    Each show PG at z=0.22 and ELG at z=0.65.
    Cross-correlation sigma is higher for each for the one indexed at 11, as is R.
    Index 12 shows a higher sigma for the source than the lens, in effect getting them backwards.
    For the above reasons I will keep Index 11.
### G262874 has two entries.
    Both show z1 to be a PG at z=0.38.
    Index 26 shows the second match at closer redshift as an ELG+PG.
    Index 27 is clearly the more reasonable one.

```python
links_dropped_duplicates = links.drop(index=[12, 26])
print(len(links_dropped_duplicates))
```

```python
#links_dropped_duplicates.reset_index(inplace=True)
links_dropped_duplicates.drop(['level_0', 'index', 'Unnamed: 0', 'Unnamed: 0.1'], axis=1, inplace=True)
links_dropped_duplicates.columns
```

```python
links_dropped_duplicates
```

```python
links_knabel[links_knabel.GAMA_ID.duplicated(keep=False)]
```

### For G544226 in links_knabel
    I'm keeping Index 1 and dropping Index 2 for the same reasons listed above.

```python
links_knabel_dropped_duplicates = links_knabel.drop(index=2)
```

```python
len(links_knabel_dropped_duplicates)
```

```python
# save each to latest
links_dropped_duplicates.to_csv(f'{latest_path}links_sample_latest_len40.csv')
links_knabel_dropped_duplicates.to_csv(f'{latest_path}links_knabel_sample_latest_len6.csv')
# save to archive
links_dropped_duplicates.to_csv(f'{csv_path}links_sample_011521_len40.csv')
links_knabel_dropped_duplicates.to_csv(f'{csv_path}links_knabel_sample_011521_len40.csv')
```
