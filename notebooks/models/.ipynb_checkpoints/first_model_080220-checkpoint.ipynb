```python
# 09/03/20 - See if this fixes the problem
from pyprojroot import here

workspace_path = str(here())
print("Workspace Path: ", workspace_path)

conf.instance = conf.Config(
    config_path="/home/shawn/Desktop/autolens_workspace/howtolens/config",
    output_path= G3629152_1933_path,
)
```

```python
imaging = al.Imaging.from_fits(
    image_path=f"{G3629152_1933_path}1933_r.fits",
    noise_map_path=f"{G3629152_1933_path}1933_r_noisemap.fits",
    psf_path=f"{G3629152_1933_path}trimmed_psf.fits",
    pixel_scales=0.198,
)
```

```python
# show the images
aplt.Imaging.subplot_imaging(imaging)
```

### It appears that it might be necessary to convolve the g,r,i images or to cut some area out (zoom in).

```python
# see what applying a mask looks like

mask = al.Mask.circular(
    shape_2d=imaging.shape_2d, pixel_scales=imaging.pixel_scales, radius=2.5, sub_size=1
)

# see what it looks like
aplt.Imaging.subplot_imaging(imaging=imaging, mask=mask)
```

```python
# specify the imaging settings and grid
settings_masked_imaging = al.SettingsMaskedImaging(grid_class=al.Grid, sub_size=2)

settings = al.SettingsPhaseImaging(settings_masked_imaging=settings_masked_imaging)
```

```python
# specify galaxy profiles and redshifts
lens_galaxy = al.GalaxyModel(redshift = G3629152_1933.zlens,
                            light = al.lp.EllipticalSersic,
                            mass = al.mp.EllipticalIsothermal)

source_galaxy = al.GalaxyModel(redshift = G3629152_1933.zsource,
                              light = al.lp.EllipticalExponential)
```

```python
# set up phase... start with 50 live points

phase = al.PhaseImaging(
    phase_name = 'G3629152_1933',
    settings = settings,
    galaxies = (lens_galaxy, source_galaxy),
    search=af.DynestyStatic(n_live_points = 50)
)
```

```python
# set up a timer for the run and store the time
from time import perf_counter as timer

time_doc = f'{G3629152_1933_path}time_doc.txt'
tic = timer()
for i in range(100):
    j = i**3
toc = timer()
time = f'{toc - tic}'
print(time)

with open(time_doc, 'w') as file_object:
    file_object.write(f"{time} seconds")
```

```python
# Run phase

time_log = f'{G3629152_1933_path}time_log.txt' # create time log
tic = timer() # start timer
print('Phase begun... hold onto ya butts.')

result = phase.run(dataset = imaging, # run phase and save output as object
                  mask = mask)

toc = timer() # stop timer
print(f'Phase complete in {toc - tic} seconds')
with open(time_doc, 'w') as file_object: # log time
    file_object.write(f"{time} seconds")
```
