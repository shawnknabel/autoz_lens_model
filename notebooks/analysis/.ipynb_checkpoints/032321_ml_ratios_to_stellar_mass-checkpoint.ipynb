```python
ml = [ [1933, 5.7031, 3.5814, 9.4309],
     [2730, 0.6208, 0.5239, 0.7065],
     [3212, 0.4801, 0.3489, 0.6642],
     [3234, 0.0001, 0.0000, 0.0117]]
ml_df = pd.DataFrame(ml, columns=['ID', 'ML', 'ML_err_low', 'ML_err_hi'])
ml_df
```

```python
samples = pd.read_csv(f'{latest_path}all_samples_latest.csv')[ ['GAMA_ID', 'ID', 'lambdar_log_mstar', 'lambdar_log_mstar_error', 'score', 'zlens', 'zsource'] ]
samples.sort_values('score', ascending=False)
```

```python
# merge those
foursome = pd.merge(samples, ml_df, how='inner', on='ID')
foursome.drop(3, inplace=True)
```

```python
# plot ml to stellar mass
mass = foursome.lambdar_log_mstar
mass_err = foursome.lambdar_log_mstar_error
ml = foursome.ML
ml_err_lo = ml - foursome.ML_err_low
ml_err_hi = foursome.ML_err_hi - ml

fig, axs = plt.subplots(figsize=(8,6))
axs.errorbar(mass, ml,
              xerr=mass_err, yerr=(ml_err_lo, ml_err_hi), 
             marker='s', linestyle='None')
```

```python
# load G3CGal.fits
group_catalog = fits.open(f'{fits_path}G3CGal.fits')[1].data
group_byte_swap=group_catalog.byteswap().newbyteorder()
groups = pd.DataFrame(group_byte_swap[:])#, columns = group_catalog.columns.names)

groups.rename(columns={'CATAID':'GAMA_ID', 'Dec':'DEC'}, inplace=True)

print(groups.columns)
#print(len(groups))

#print(groups.head())

#print(group_byte_swap[:5])
```

```python
foursome = pd.merge(foursome, groups, how='left', on='GAMA_ID')
foursome
```

```python
# plot ml to stellar mass
ID = foursome.ID
group = ['N/A', 'Isolated', 'Group']
mass = foursome.lambdar_log_mstar
mass_err = foursome.lambdar_log_mstar_error
ml = foursome.ML
ml_err_lo = ml - foursome.ML_err_low
ml_err_hi = foursome.ML_err_hi - ml

print(ml_err_lo[1])

fig, axs = plt.subplots(figsize=(8,6))
for i in range(3):
    axs.errorbar(mass[i], ml[i],
                 xerr=mass_err[i], yerr=([ml_err_hi[i]], [ml_err_hi[i]]),
                marker='s', linestyle='None', label=f'{int(ID[i])} - {str(group[i])}')
axs.legend(loc='lower left')
axs.set_title('Mass to Light Ratio vs. Stellar Mass')
axs.set_xlabel('Stellar Mass $\log(M_*/M_{\odot})$')
axs.set_ylabel('Mass to Light Ratio')
plt.savefig(f'{png_path}ml_ratio_stellar_mass_032321.png')
```
