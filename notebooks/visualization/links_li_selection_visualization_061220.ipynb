# This notebook visualizes sigma selection of Links and Li.

```python
# libraries
import numpy as np
import pandas as pd
from astropy.io import fits
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
```

```python
# read in catalogs
links = pd.read_csv('/home/shawn/Desktop/links_autoz/files/csv/links_autoz052220.csv')
links_identified = pd.read_csv('/home/shawn/Desktop/links_autoz/files/csv/links_identified_autoz052220.csv')
links_select= pd.read_csv('/home/shawn/Desktop/links_autoz/files/csv/links_select052920.csv')
links_identified_select = pd.read_csv('/home/shawn/Desktop/links_autoz/files/csv/links_identified_select052920.csv')
li = pd.read_csv('/home/shawn/Desktop/links_autoz/files/csv/li_autoz_latest.csv')
li_select = pd.read_csv('/home/shawn/Desktop/links_autoz/files/csv/li_select_latest.csv')
```

```python
# plots

fig = plt.figure(figsize=(8,6))
ax = plt.subplot()
ax.set_xlim(0.9, 2.5)
ax.set_ylim(2.0, 10.5)
ax.scatter(links.R,
          links.sigma2,
          color = 'k',
          marker = '.',
          alpha = 0.75,
          label = 'LinKS in GAMA')
ax.scatter(links_identified.R,
           links_identified.sigma2,
          color = 'g',
          marker = 'o',
          alpha = 0.75,
          label = 'LinKS Candidates from Knabel-2020')
ax.scatter(links_select.R,
          links_select.sigma2,
          facecolor= 'none',
          marker = 's',
          edgecolor = 'r')
ax.scatter(li.R,
           li.sigma2,
          color = 'orange',
          marker = 'o',
          alpha = 0.75,
          label = 'Li Candidates')
ax.scatter(li_select.R,
          li_select.sigma2,
          facecolor= 'none',
          marker = 's',
          edgecolor = 'r',
          label = 'Selected Candidates')
ax.add_patch(patches.Rectangle((1.85, 4.5), 2, 10, 
                               edgecolor = 'k', facecolor = 'none', linestyle = ':',
                              label = 'Holwerda-15 Selection'))
ax.axvline(1.2, color = 'k', linewidth = .75, linestyle = '--', label = 'R = 1.2')
ax.set_xlabel('$R = \sigma_2/\sqrt{(\sigma_3)^2/2+(\sigma_4)^2/2}$', fontsize = 12)
ax.set_ylabel('$\sigma_2$',  fontsize = 12)
ax.set_title('Selection Based on $\sigma$ and R Values')
plt.legend(fontsize=9)
#plt.show()

#plt.savefig('/home/shawn/Desktop/links_autoz/visuals/png/sigma_selection062020.png')
```
