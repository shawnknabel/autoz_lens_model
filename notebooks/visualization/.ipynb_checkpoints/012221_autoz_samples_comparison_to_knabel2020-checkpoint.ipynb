# Einstein Radius Estimates

## Point mass:
### $\theta_E = (M_{r_e/2}/(10^{8.09}M_{\odot}))^{1/2}(D_{LS}/(D_LD_S) Mpc)^{1/2}arcsec$
## SIS:
### $\theta_E = (\sigma_{SIS}/(186km/s))^2(D_{LS}/D_S) arcsec$

## Half-effective radius mass
## $M_{r_e/2} = 0.0011(M_*)^{1.25}$

## Velocity Dispersion Estimates
## $\sigma(M_*) = \sigma_b * (M_* / M_b)^{\alpha}$
### For z < 0.2: (SDSS fit)
#### $\sigma = 10^{2.073} (M_*/ 10^{10.26} M_{\odot})^{\alpha}$
#### where $\alpha = 0.403$ if $M_* <= 10^{10.26}$, or $\alpha = 0.293$ otherwise
### For z > 0.2: (SHELS fit)
#### $\sigma = 10^{2.071} (M_* / 10^{10.26} M_{\odot})^{0.281}$

```python
# add half-effective radius masses from Auger2010
links['m_half_re'] = 0.0011 * (links.lambdar_mstar) ** (1.25)
links_knabel['m_half_re'] = 0.0011 * (links_knabel.lambdar_mstar) ** (1.25)
li['m_half_re'] = 0.0011 * (li.lambdar_mstar) ** (1.25)
```

```python
# create velocity disperions estimator
def calculate_sigma (z, mass):
    mb = 10**(10.26)
    if z < 0.2: # SDSS fit
        sigma_b = 10**(2.073)
        if mass < mb:
            alpha = 0.403
        else:
            alpha = 0.293
        sigma = sigma_b * ( mass / mb ) ** (alpha)
    elif (z >= 0.2) & (z < 0.7): # SHELS fit
        sigma_b = 10**(2.071)
        alpha = 0.281
        sigma = sigma_b * ( mass / mb ) ** (alpha)
    else:
        print(f'Redshift out of range. {z}')
        sigma_b = 10**(2.071)
        alpha = 0.281
        sigma = sigma_b * ( mass / mb ) ** (alpha)
    return sigma

# define function to calculate from dataframe
def calculate_sigmas (z, mass):
    sigmas = []
    for i in range(len(z)):
        sigma = calculate_sigma(z[i], mass[i])
        sigmas.append(sigma)
    return sigmas
```

```python
# add sigma-star
links['sigma_star'] = calculate_sigmas(z= links.zlens, mass = links.lambdar_mstar)
links_knabel['sigma_star'] = calculate_sigmas(z= links_knabel.zlens, mass = links_knabel.lambdar_mstar)
li['sigma_star'] = calculate_sigmas(z= li.zlens, mass = li.lambdar_mstar)

print(links.columns, links_knabel.columns, li.columns)
print(links['sigma_star'])
```

```python
# EINSTEIN RADIUS ESTIMATES

def theta_e_estimates(sample):
    sample['theta_e_pm']=np.zeros(len(sample))
    sample['theta_e_sis']=np.zeros(len(sample))
    for i in range(len(sample)):
        M = sample.m_half_re.iloc[i]
        sigma = sample.sigma_star[i]
        DL = sample.dlens.iloc[i]
        DS = sample.dsource.iloc[i]
        sample['theta_e_pm'].iloc[i] = (M/(10**(8.09)))**(1/2) * ((DS - DL)/(DL*DS))**(1/2)
        sample['theta_e_sis'].iloc[i] = (sigma / 186)**2 * (DS - DL)/DS
    print('Estimates estimated!')
    return sample[['theta_e_pm','theta_e_sis']]
```

```python
# estimates on autoz samples
theta_e_estimates(links)
theta_e_estimates(links_knabel)
theta_e_estimates(li)
#print(links[['theta_e_pm','theta_e_sis']], links_knabel[['theta_e_pm','theta_e_sis']], li[['theta_e_pm','theta_e_sis']])
```
